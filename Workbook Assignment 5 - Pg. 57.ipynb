```python
import pandas as pd
names = ['Bob','Jessica','Mary','John','Mel']
grades = [76,95,77,78,99]
bsdegrees = [1,1,0,0,1]
msdegrees = [2,1,0,0,0]
phddegrees = [0,1,0,0,0]
EducationList = zip(names,grades,bsdegrees,msdegrees,phddegrees)
df = pd.DataFrame(data=EducationList,
                 columns=['Names','Grades','BS Degrees','MS Degrees','PhD Degrees'])
df
```

```python
#Summary Statistics of Grades
gradeCount = df['Grades'].count() # number of values
print("The number of grades in the dataset is:", gradeCount)

gradeMean = df['Grades'].mean() # arithmetic average
print("The average grade is:", gradeMean)

gradeMedian = df['Grades'].median() #median
print("The median grade is:", gradeMedian)

gradeMode = df['Grades'].mode() #mode
print("The mode for grades is:")
print(gradeMode)

gradeVariance = df['Grades'].var() #variance
print("The variance for grades is:", gradeVariance)

gradeStd = df['Grades'].std() # standard deviation
print("The standard deviation for grades is:", gradeStd)

gradeMin = df['Grades'].min() # minimum
print("The minimum grade is:", gradeMin)

gradeMax = df['Grades'].max() # maximum
print("The maximum grade is:", gradeMax)

gradeFirstQuantile = df['Grades'].quantile(.25) # first quartile
print("The first quantile (.25) for grades is:", gradeFirstQuantile)

gradeSecondQuantile = df['Grades'].quantile(.5) # second quartile
print("The second quantile (.50) for grades is:", gradeSecondQuantile)

gradeThirdQuantile = df['Grades'].quantile(.75) # third quartile
print("The third quantile (.75) for grades is:", gradeThirdQuantile)
```

```python
#Summary Statistics of BS Degrees
bsDegreesCount = df['BS Degrees'].count() # number of values
print("The number of BS degrees in the dataset is:", bsDegreesCount)

bsDegreesMean = df['BS Degrees'].mean() # arithmetic average
print("The average number of BS degrees is:", bsDegreesMean)

bsDegreesMedian = df['BS Degrees'].median() #median
print("The median for BS degrees is:", bsDegreesMedian)

bsDegreesMode = df['BS Degrees'].mode() #mode
print("The mode for BS degrees is:")
print(bsDegreesMode)

bsDegreesVariance = df['BS Degrees'].var() #variance
print("The variance for BS degrees is:", bsDegreesVariance)

bsDegreesStd = df['BS Degrees'].std() # standard deviation
print("The standard deviation for BS degrees is:", bsDegreesStd)

bsDegreesMin = df['BS Degrees'].min() # minimum
print("The minimum for BS degrees is:", bsDegreesMin)

bsDegreesMax = df['BS Degrees'].max() # maximum
print("The maximum for BS degrees is:", bsDegreesMax)

bsDegreesFirstQuantile = df['BS Degrees'].quantile(.25) # first quartile
print("The first quantile (.25) for BS degrees is:", bsDegreesFirstQuantile)

bsDegreesSecondQuantile = df['BS Degrees'].quantile(.50) # second quartile
print("The second quantile (.50) for BS degrees is:", bsDegreesSecondQuantile)

bsDegreesThirdQuantile = df['BS Degrees'].quantile(.75) # third quartile
print("The third quantile (.75) for BS degrees is:", bsDegreesThirdQuantile)
```
